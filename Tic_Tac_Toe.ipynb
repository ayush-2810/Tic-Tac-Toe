```python
print('Welcome To Tic Tac Toe.')


def showb():
    print('',board[0],'|',board[1],'|',board[2],'|')
    print('-------------')
    print('',board[3],'|',board[4],'|',board[5],'|')
    print('-------------')
    print('',board[6],'|',board[7],'|',board[8],'|')

    
board=[1,2,3,4,5,6,7,8,9]
k1=input('Enter first player name: ')
k2=input('Enter second player name: ')
print()
print(k1,' = X',sep='')
print(k2,' = O',sep='')
print()
showb()


def check(c,p1,p2,p3):
    if board[p1]==c and board[p2]==c and board[p3]==c:
        return True

    
def check1(c):
    if check(c,0,1,2):
        return True
    if check(c,3,4,5):
        return True
    if check(c,6,7,8):
        return True
    if check(c,0,4,8):
        return True
    if check(c,0,3,6):
        return True
    if check(c,1,4,7):
        return True
    if check(c,2,5,8):
        return True
    if check(c,2,4,6):
        return True

    
def p1():
    while True:
        print()
        print(k1,'move: ')
        n=int(input())
        print()
        if n>10 or n<1 or board[n-1]=='X' or board[n-1]=='O' :
            print('INVALID MOVE.')
        else:
            board[n-1]='X'
            showb()
            break

            
def p2():
    while True:
        print()
        print(k2,' move:')
        n=int(input())
        print()
        if n>10 or n<1 or board[n-1]=='O' or board[n-1]=='X' :
            print('INVALID MOVE.')
        else:
            board[n-1]='O'
            showb()
            break
        

def game():
    flag=True
    while flag:
        p1()
        
        x=board.count('X')
        o=board.count('O')
    
        if check1('X')==True:
            print()
            flag=False
            print('~~~',k1,' wins~~~',sep='')
            break
        
        if x+o==9:
            print()
            print('Tie')
            break
        
        
        p2()
        
        x=board.count('X')
        o=board.count('O')
    
        if check1('O')==True:
            print()
            flag=False
            print('~~~',k2,' wins~~~',sep='')
            break
        
        if x+o==9:
            print()
            print('Tie')
        print()
        
game()
```
