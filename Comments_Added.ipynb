```python
print('Welcome To Tic Tac Toe.')   


def showb():   #Function for showing the board.
    print('',board[0],'|',board[1],'|',board[2],'|')
    print('-------------')
    print('',board[3],'|',board[4],'|',board[5],'|')
    print('-------------')
    print('',board[6],'|',board[7],'|',board[8],'|')

    
board=[1,2,3,4,5,6,7,8,9]   #Initializing the board with a list.
k1=input('Enter first player name: ')   #Taking input for player names.
k2=input('Enter second player name: ')
print()
print(k1,' = X',sep='')
print(k2,' = O',sep='')
print()
showb()   #Calling function for showing the board.


def check(c,p1,p2,p3):   #Function for checking whether all the squares have same character or not for winning.
    if board[p1]==c and board[p2]==c and board[p3]==c:
        return True

    
def check1(c):   #Function for checking which all squares will elegible for winning.
    if check(c,0,1,2):   #Checking for particular squares.
        return True
    if check(c,3,4,5):
        return True
    if check(c,6,7,8):
        return True
    if check(c,0,4,8):
        return True
    if check(c,0,3,6):
        return True
    if check(c,1,4,7):
        return True
    if check(c,2,5,8):
        return True
    if check(c,2,4,6):
        return True

    
def p1():   #Function for making player 1 moves.
    while True:
        print()
        print(k1,'move: ')
        n=int(input())
        print()
        if n>10 or n<1 or board[n-1]=='X' or board[n-1]=='O' :   #For checking whether the square is already filled or not.
            print('INVALID MOVE.')
        else:
            board[n-1]='X'   #Assigning value for a square.
            showb()
            break

            
def p2():   #Function for making player 2 moves.
    while True:
        print()
        print(k2,' move:')
        n=int(input())
        print()
        if n>10 or n<1 or board[n-1]=='O' or board[n-1]=='X' :   #For checking whether the square is already filled or not.
            print('INVALID MOVE.')
        else:
            board[n-1]='O'   #Assigning value for a square.
            showb()
            break
        

def game():   #Function for starting the game.
    flag=True
    while flag:
        p1()   #Calling function for player 1 move.
        
        x=board.count('X')
        o=board.count('O')
    
        if check1('X')==True:
            print()
            flag=False
            print('~~~',k1,' wins~~~',sep='')
            break
        
        if x+o==9:   #Condition for checking for tie.
            print()
            print('Tie')
            break
        
        
        p2()   #Calling function for player 1 move.
        
        x=board.count('X')
        o=board.count('O')
    
        if check1('O')==True:
            print()
            flag=False
            print('~~~',k2,' wins~~~',sep='')
            break
        
        if x+o==9:   #Condition for checking for tie.
            print()
            print('Tie')
        print()
        
game()   #Calling function for starting the game.
```
